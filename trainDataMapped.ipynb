```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
trainData = pd.read_csv("train.csv",low_memory = False)
```

```python
trainData = pd.read_csv("test.csv",low_memory = False)
```

```python
trainData.head()
```

```python
trainData["Fare_mean"] = trainData["Fare"] < trainData["Fare"].mean()
```

```python
def findWord(word,string):
    if string.find(word) != -1:
        return True
    return False
```

```python
trainData["Mr"] = trainData["Name"].map(lambda x: findWord("Mr.",x))
trainData["Mrs"] = trainData["Name"].map(lambda x: findWord("Mrs.",x))
trainData["Miss"] = trainData["Name"].map(lambda x: findWord("Miss.",x))
trainData["Master"] = trainData["Name"].map(lambda x: findWord("Master.",x))
trainData["Dr"] = trainData["Name"].map(lambda x: findWord("Dr.",x))
trainData["Don"] = trainData["Name"].map(lambda x: findWord("Don.",x))
trainData["Rev"] = trainData["Name"].map(lambda x: findWord("Rev.",x))
trainData["Major"] = trainData["Name"].map(lambda x: findWord("Major.",x))
trainData["Mme"] = trainData["Name"].map(lambda x: findWord("Mme.",x))
trainData["Ms"] = trainData["Name"].map(lambda x: findWord("Ms.",x))
trainData["Lady"] = trainData["Name"].map(lambda x: findWord("Lady.",x))
trainData["Sir"] = trainData["Name"].map(lambda x: findWord("Sir.",x))
trainData["Mlle"] = trainData["Name"].map(lambda x: findWord("Mlle.",x))
trainData["Col"] = trainData["Name"].map(lambda x: findWord("Col.",x))
trainData["Capt"] = trainData["Name"].map(lambda x: findWord("Capt.",x))
trainData["Countess"] = trainData["Name"].map(lambda x: findWord("Countess.",x))
trainData["Jonkheer"] = trainData["Name"].map(lambda x: findWord("Jonkheer.",x))
```

```python
trainData["OtherTitles"] = trainData["Don"] | trainData["Rev"] | trainData["Major"]| trainData["Mme"]| trainData["Ms"] | trainData["Lady"] |trainData["Sir"]| trainData["Mlle"]| trainData["Col"]|trainData["Capt"]|trainData["Countess"] | trainData["Jonkheer"]
```

```python
trainData["Sex_female"] = trainData["Sex"] == "female"
```

```python
trainData["Pclass_3"] = trainData["Pclass"] == 3
trainData["Pclass_2"] = trainData["Pclass"] == 2
trainData["Pclass_1"] = trainData["Pclass"] == 1
```

```python
trainData.head()
```

```python
trainData["Embarked_C"] = trainData["Embarked"] == "C"
trainData["Embarked_Q"] = trainData["Embarked"] == "Q"
trainData["Embarked_S"] = trainData["Embarked"] == "S"
```

```python
trainData.head()
```

```python
def mapTicket(x):
    if(len(x.split(" ")) == 2):
        return x.split(" ")[1]
    elif(len(x.split(" ")) == 3):
        return x.split(" ")[2]
    elif(str(x).isdigit()):
        return x
    elif x == "LINE":
        return "500"
```

```python
trainData["Ticket"] = trainData[(trainData["Ticket"].map(lambda x: mapTicket(x))).astype(str).str.isdigit() == True]["Ticket"].map(lambda x: mapTicket(x))
```

```python
trainFinal = trainData[['Survived', 'Age', 'SibSp','Cabin',"Ticket",
       'Parch', 'Fare', 'Sex_female', 'Pclass_3',
       'Pclass_2', 'Pclass_1', 'Embarked_C', 'Embarked_Q', 'Embarked_S']]
```

```python
trainFinal["Age"] = trainFinal["Age"].fillna(trainFinal["Age"].mean())
```
